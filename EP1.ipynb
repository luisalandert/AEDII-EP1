## Funcao para processar os dados

```python
def processa_df (df_entrada, df_saida):
    #formata df_saida
    df_saida.loc[len(df_saida)] = [df_entrada.iloc[0,0]]
    df_saida['frequentadores'] = np.empty((len(df_saida), 0)).tolist()
    (df_saida.iloc[0,1]).append(df_entrada.iloc[0,1])
    #processamento dos dados
    cont = 0
    indice_coord = 0
    for item in range(len(df_entrada)-1):
        coord_atual = df_entrada.iloc[cont, 0] #linha cont coluna 0 (celula da coordenada)
        coord_prox = df_entrada.iloc[cont+1, 0] #coordenada da proxima linha
        if coord_prox == coord_atual:
            (df_saida.iloc[indice_coord, 1]).append(df_entrada.iloc[cont+1, 1])#colocando o id_pess na lista frequentadores
        else:
            df_saida.loc[len(df_saida)] = [df_entrada.iloc[cont+1, 0], [df_entrada.iloc[cont+1, 1]]]
            indice_coord += 1
        cont += 1
```
